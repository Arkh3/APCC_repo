# Teleportación cuántica en Qiskit.

## Parte 1: Simulación.

```python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit import IBMQ, Aer, transpile, assemble
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.extensions import Initialize
from qiskit_textbook.tools import random_state, array_to_latex
```

```python
def crear_estado_bell(circuit, cubit_a, cubit_b):
    circuit.barrier()
    circuit.h(cubit_a)
    circuit.cx(cubit_a,cubit_b)
    
def parte_de_alice(circuit, psi, cubit_a):
    circuit.barrier()
    circuit.cx(psi, cubit_a)
    circuit.h(psi)
    circuit.barrier()
    circuit.measure(psi,0)
    circuit.measure(cubit_a,1)

def parte_de_bob(circuit, cubit, crz, crx):
    circuit.barrier()
    circuit.x(cubit).c_if(crx, 1)
    circuit.z(cubit).c_if(crz, 1)
```

```python
# Creamos un estado aleatorio de un cubit
psi = random_state(1)


array_to_latex(psi, pretext="|\\psi\\rangle =")
plot_bloch_multivector(psi)
```

```python
init_gate = Initialize(psi)
init_gate.label = "init"
```

```python
qr = QuantumRegister(3, name="q")      # El protocolo usa 3 cubits
crz = ClassicalRegister(1, name="crz") # y dos bits clásicos
crx = ClassicalRegister(1, name="crx")
qc = QuantumCircuit(qr, crz, crx)

# Paso 0
# Inicializar el cubit de Alice a un estado aleatorio
qc.append(init_gate, [0])

# Paso 1
# Crear el estado de bell
crear_estado_bell(qc, 1, 2)

## STEP 2
# 
parte_de_alice(qc, 0, 1)

## STEP 4
# Bob decodes qubits
parte_de_bob(qc, 2, crz, crx)

# Display the circuit
qc.draw()
```

-------------------

## Simulación con Statevector:

```python
sv_sim = Aer.get_backend('statevector_simulator')
qobj = assemble(qc)
out_vector = sv_sim.run(qobj).result().get_statevector()
plot_bloch_multivector(out_vector)
```

-------------------
Como podemos ver, los cubits 0 y 1 han colapsado perdiendo su estado y el cubit 2 ha adquirido el estado que tenía el cubit 0 al principio del protocolo (está un par de celdas arriba).

No puedo añadir muchos más comentarios que en el ejercicio 1 en Cirq pero aquí se pueden visualizar mucho mejor los estados de los cubits y como colapsan.

## Simulación con QSAM:

La forma en la que comprobaremos que el protocolo esta simulando en un simulador QSAM (en el que no podemos ver el vector de estados), sería empezando con un cubit a en el estado |0> y inicializarlo a un estado. 

Como las operaciones cuánticas tienen que ser reversibles, significa que si a el nuevo estado le aplicamos las operaciones cuánticas para revertirlo, este debería volver al estado |0>. 

Si inicializamos el cubit a un estado, se lo mandamos a Bob, y al cubit de Bob le aplicamos las operaciones para revertir el estado y esto nos devuelve un cubit con estado |0>, habremos demostrado que la teleportación cuántica funciona.

```python
# operaciones inversas a la inicialización
inverse_init_gate = init_gate.gates_to_uncompute()
```

```python
qr = QuantumRegister(3, name="q")      
crz = ClassicalRegister(1, name="crz") 
crx = ClassicalRegister(1, name="crx")
qc = QuantumCircuit(qr, crz, crx)

qc.append(init_gate, [0])
qc.barrier()

crear_estado_bell(qc, 1, 2)

parte_de_alice(qc, 0, 1)

parte_de_bob(qc, 2, crz, crx)

qc.append(inverse_init_gate, [2])

qc.draw()
```

```python
# Necesitamos un nuevo bit clásico para ver el resultado del
# cubit mandado a Bob tras el teletransporte cuántico.
cr_result = ClassicalRegister(1)
qc.add_register(cr_result)
qc.measure(2,2)
qc.draw()
```

```python
# Ejecutamos el circuito
qasm_sim = Aer.get_backend('qasm_simulator')
t_qc = transpile(qc, qasm_sim)
qobj = assemble(t_qc)
counts = qasm_sim.run(qobj).result().get_counts()
plot_histogram(counts)
```

Como se puede ver en el histograma, tenemos una probabilidad del 100% de que el cubit con el resultado colapse a 0 (lo que significa que está en el estado |0>) y probando así que antes de hacer las operaciones invertidas, Bob tenía un cubit con el mismo estado que el cubit original de Alice.

## Parte 2: Ejecución en backends reales:

En los ordenadores cuánticos de IBM no se pueden realizar mas instrucciones tras hacer las mediciones de los cubits.

Esto hacer que tengamos que cambiar el código ya que hay un principio que nos dice que podemos posponer todas las mediciones de los cubits a el final y los resultados deberían ser los mismos.

Esta solución nos permite demonstrar la teleportación cuántica en un ordenador cuántico real pero perdemos la capacidad de transmitir estados cuánticos mediante canales de comunicación clásicos.

```python
def parte_bob_2(qc, a, b, c):
    qc.barrier()
    qc.cx(b, c)
    qc.cz(a, c)
```

```python
def parte_de_alice_2(circuit, psi, cubit_a):
    circuit.barrier()
    circuit.cx(psi, cubit_a)
    circuit.h(psi)
```

```python
qc = QuantumCircuit(3,1)

qc.append(init_gate, [0])

crear_estado_bell(qc, 1, 2)

parte_de_alice_2(qc, 0, 1)

parte_bob_2(qc, 0, 1, 2)

qc.append(inverse_init_gate, [2])

qc.measure(2,0)

qc.draw()
```
