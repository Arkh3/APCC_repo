# Sumador de Draper con la QFT

```python
from pytket import Circuit
from pytket.backends.ibm import AerStateBackend, AerBackend
from qiskit.visualization import plot_bloch_multivector

c = Circuit(8,8)  # define a circuit with 4 qubits and 4 bits

##inicializo A a |1101>
c.X(3)
c.X(2)
c.X(0)
##inicializo B a |0001>
c.X(7)
c.add_barrier([0, 1, 2, 3,4,5,6,7])

##QFT

c.H(3) 
c.CRz(0.5, 2, 3) 
c.CRz(0.25, 1, 3)
c.CRz(0.125, 0, 3)
c.add_barrier([0, 1, 2, 3,4,5,6,7])


c.H(2) 
c.CRz(0.5, 1, 2)
c.CRz(0.25, 0, 2)
c.add_barrier([0, 1, 2, 3])

c.H(1) 
c.CRz(0.5, 0, 1)
c.add_barrier([0, 1, 2, 3])

c.H(0)

c.SWAP(0,3)
c.SWAP(1,2)

##Adder

c.CRz(0.125, 7, 0)
c.CRz(0.25, 6, 0)
c.CRz(0.5, 5, 0)
c.CRz(1, 4, 0)
c.add_barrier([0, 1, 2, 3,4,5,6,7])

c.CRz(0.25, 7, 1)
c.CRz(0.5, 6, 1)
c.CRz(1, 5, 1)
c.add_barrier([0, 1, 2, 3,4,5,6,7])

c.CRz(0.5, 7, 2)
c.CRz(1, 6, 2)
c.add_barrier([0, 1, 2, 3,4,5,6,7])

c.CRz(1, 7, 3)
c.add_barrier([0, 1, 2, 3,4,5,6,7])


##QFT inversa

c.SWAP(0,3)
c.SWAP(1,2)

c.add_barrier([0, 1, 2, 3,4,5,6,7])
c.H(0)

c.add_barrier([0, 1, 2, 3])
c.CRz(-0.5, 0, 1)
c.H(1) 

c.add_barrier([0, 1, 2, 3])
c.CRz(-0.25, 0, 2)
c.CRz(-0.5, 1, 2)
c.H(2) 

c.add_barrier([0, 1, 2, 3])
c.CRz(-0.125, 0, 3)
c.CRz(-0.25, 1, 3)
c.CRz(-0.5, 2, 3) 
c.H(3) 

c.measure_all()
b = AerBackend()
b.compile_circuit(c)
handle = b.process_circuit(c, 1000)
result = b.get_result(handle)
counts = result.get_counts()
print(counts)
```

Nos da el resultado en los 4 bits mas significativos,nos lo da bien pero invertido (0001 + 1101 = 1110)y a nosotros nos da 0111.
