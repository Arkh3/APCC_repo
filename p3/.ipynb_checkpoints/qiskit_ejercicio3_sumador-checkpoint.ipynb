```python
def qft_4qbits(qc):

    qc.h(3)
    qc.cp(pi/2,2,3)
    qc.cp(pi/4,1,3)
    qc.cp(pi/8,0,3)

    qc.h(2)
    qc.cp(pi/2,1,2)
    qc.cp(pi/4,0,2)

    qc.h(1)
    qc.cp(pi/2,0,1)

    qc.h(0)
          
    qc.swap(0,3)
    qc.swap(1,2)
    
    
def inverse_qft_4qbits(qc):
    
    qc.swap(0,3)
    qc.swap(1,2)
    
    qc.h(0)
    
    qc.cp(-pi/2,0,1)
    qc.h(1)
    
    qc.cp(-pi/4,0,2)
    qc.cp(-pi/2,1,2)
    qc.h(2)
    
    qc.cp(-pi/8,0,3)
    qc.cp(-pi/4,1,3)
    qc.cp(-pi/2,2,3)  
    qc.h(3)

    
def adder(qc):
    
    qc.cp(pi/8,7,0)
    qc.cp(pi/4,6,0)
    qc.cp(pi/2,5,0)
    qc.cp(pi,4,0)
    
    qc.barrier()
    
    qc.cp(pi/4,7,1)
    qc.cp(pi/2,6,1)
    qc.cp(pi,5,1)
    
    qc.barrier()
    
    qc.cp(pi/2,7,2)
    qc.cp(pi,6,2)
    
    qc.barrier()
    
    qc.cp(pi,7,3)
    
    qc.barrier()
```

def adder(qc):
    
    qc.cp(pi,4,0)
    
    qc.barrier()

    qc.barrier()
    
    qc.cp(pi,6,2)

    qc.barrier()
    
    qc.cp(pi,7,3)

```python
qc = QuantumCircuit(8,8) #4qbits para A y 4 para B

#inicializo A a |1101> (13 en binario)
qc.x(3)
qc.x(2)
qc.x(0)

#inicializo B a |0001> (1 en binario)
qc.x(4)

qc.barrier()
qft_4qbits(qc)

qc.barrier()
adder(qc)

qc.barrier()
inverse_qft_4qbits(qc)

qc.barrier()
qc.measure([0,1,2,3,4,5,6,7],[0,1,2,3,4,5,6,7])

qc.draw()
```

```python
sim = Aer.get_backend("qasm_simulator")
job = execute(qc,sim,shots=1000)
result = job.result()
counts = result.get_counts(qc)

print(counts)
```
