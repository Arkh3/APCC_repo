# ALGORITMO DE DEUTSCH-JOZSA

Ahora programaremos el algoritmo de Deutsch-Jozsa en Qiskit.

----

## 1- Crearemos todos los oráculos y los guardamos en un diccionario.

Hay que crear:

a) Oráculos de 2 bits:
- 2 constantes
- 2 balanceados

b) Oráculos de 3 bits:
- 2 constantes 
- 2 balanceados

### * Oráculos de 2 bits:

```python
# oráculo constante de 2 bits
n = 2

const_oracle0_2b = QuantumCircuit(n+1)

const_oracle0_2b.draw()
```

```python
# oráculo constante de 2 bits

const_oracle1_2b = QuantumCircuit(n+1)
const_oracle1_2b.x(n)

const_oracle1_2b.draw()
```

```python
# oráculo balanceado de 2 bits

balanced_oracle0_2b = QuantumCircuit(n+1)
balanced_oracle0_2b.cnot(0,2)
balanced_oracle0_2b.cnot(1,2)

balanced_oracle0_2b.draw()
```

```python
# oráculo balanceado de 2 bits

balanced_oracle1_2b = QuantumCircuit(n+1)
balanced_oracle1_2b.cnot(0,2)

balanced_oracle1_2b.draw()
```

### * Oráculos de 3 bits:

```python
# oráculo constante de 3 bits
n = 3

const_oracle0_3b = QuantumCircuit(n+1)

const_oracle0_3b.draw()
```

```python
# oráculo constante de 3 bits

const_oracle1_3b = QuantumCircuit(n+1)
const_oracle1_3b.x(n)

const_oracle1_3b.draw()
```

```python
balanced_oracle0_3b = QuantumCircuit(n+1)
balanced_oracle0_3b.cnot(0,3)
balanced_oracle0_3b.cnot(1,3)
balanced_oracle0_3b.cnot(2,3)

balanced_oracle0_3b.draw()
```

```python
balanced_oracle1_3b = QuantumCircuit(n+1)
balanced_oracle1_3b.cnot(0,3)

balanced_oracle1_3b.draw()
```

```python
oraculos = dict()
oraculos['const_oracle0_2b'] = const_oracle0_2b
oraculos['const_oracle1_2b'] = const_oracle1_2b
oraculos['balanced_oracle0_2b'] = balanced_oracle0_2b
oraculos['balanced_oracle1_2b'] = balanced_oracle1_2b

oraculos['const_oracle0_3b'] = const_oracle0_3b
oraculos['const_oracle1_3b'] = const_oracle1_3b
oraculos['balanced_oracle0_3b'] = balanced_oracle0_3b
oraculos['balanced_oracle1_3b'] = balanced_oracle1_3b
```

## 2- Creamos el resto del circuito del algoritmo de Deutsch-Jozsa para funciones de 2 bits

```python
def deutsch_jozsa(n,oraculo):

    dj_circuit = QuantumCircuit(n+1, n)

    # Apply H-gates
    for qubit in range(n):
        dj_circuit.h(qubit)

    # Put qubit in state |->
    dj_circuit.x(n)
    dj_circuit.h(n)
    
    dj_circuit.barrier()

    # Add oracle
    dj_circuit += oraculos[oraculo]

    dj_circuit.barrier()
    
    # Repeat H-gates
    for qubit in range(n):
        dj_circuit.h(qubit)
    dj_circuit.barrier()


    # Display circuit
    dj_circuit.draw()
    
    return dj_circuit
```

## 3- Simulamos los oráculos constantes de 2 bits

Al ser funciones constantes de 2 bits, los primeros 2 cubits deberían colapsar al estado |0> después de medirlos.

```python
#oráculo constante 1 de 2 bits
qc = deutsch_jozsa(2,"const_oracle0_2b")
qc.draw()
```

```python
sv_sim = Aer.get_backend("statevector_simulator")
qobj = assemble(qc)
statevector = sv_sim.run(qobj).result().get_statevector()
plot_bloch_multivector(statevector)
```

### Comentarios:



```python
# Measure
for i in range(2):
    qc.measure(i, i)
    
qc.draw()
```

```python
# use local simulator
qasm_sim = Aer.get_backend('qasm_simulator')
shots = 1024
qobj = assemble(qc, qasm_sim)
results = qasm_sim.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer)
```

### Comentarios:





## Simulación en un backend real:

```python
# Load our saved IBMQ accounts and get the least busy backend device with greater than or equal to (n+1) qubits
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q')
backend = least_busy(provider.backends(filters=lambda x: x.configuration().n_qubits >= (n+1) and
                                   not x.configuration().simulator and x.status().operational==True))
print("least busy backend: ", backend)
```

```python
# Run our circuit on the least busy backend. Monitor the execution of the job in the queue
from qiskit.tools.monitor import job_monitor

shots = 1024
transpiled_dj_circuit = transpile(qc, backend, optimization_level=3)
qobj = assemble(transpiled_dj_circuit, backend)
job = backend.run(qobj)
job_monitor(job, interval=2)
```

```python
# Get the results of the computation
results = job.result()
answer = results.get_counts()

plot_histogram(answer)
```
