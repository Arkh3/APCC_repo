# ALGORITMO DE DEUTSCH

Empezaremos probando el algoritmo de Deutsch.

----

## 1- Crearemos todos los oráculos posibles y los guardamos en un diccionario.

```python
oraculos = dict()
```

```python
oraculo_constante0 = QuantumCircuit(2,2)
oraculo_constante0.barrier()
oraculo_constante0.barrier()
oraculo_constante0.draw()
```

```python
oraculo_constante1 = QuantumCircuit(2,2)
oraculo_constante1.barrier()
oraculo_constante1.x(1)
oraculo_constante1.barrier()
oraculo_constante1.draw()
```

```python
oraculo_balanceado0 = QuantumCircuit(2,2)
oraculo_balanceado0.barrier()
oraculo_balanceado0.cnot(0,1)
oraculo_balanceado0.barrier()
oraculo_balanceado0.draw()
```

```python
oraculo_balanceado1 = QuantumCircuit(2,2)
oraculo_balanceado1.barrier()
oraculo_balanceado1.cnot(0,1)
oraculo_balanceado1.x(1)
oraculo_balanceado1.barrier()
oraculo_balanceado1.draw()
```

```python
oraculos['oraculo_constante0'] = oraculo_constante0
oraculos['oraculo_constante1'] = oraculo_constante1
oraculos['oraculo_balanceado0'] = oraculo_balanceado0
oraculos['oraculo_balanceado1'] = oraculo_balanceado1
```

## 2- Creamos el resto del circuito del algoritmo de Deutsch

```python
def algoritmoDeutsch(oraculo):
    
    qc = QuantumCircuit(2,2)
    
    qc.x(1)
    qc.barrier()
    qc.h(0)
    qc.h(1)
    qc = qc + oraculos[oraculo]
    qc.h(0)
    #qc.measure([0,1],[0,1])
    return qc
```

## 3- Simulamos los oráculos constantes 

Al ser funciones constantes, el cubit 0 debería colapsar en el estado |0> después de medirlo.

## Oráculo constante 1:

```python
qc = algoritmoDeutsch("oraculo_constante0")
qc.draw()
```

```python
sv_sim = Aer.get_backend("statevector_simulator")
qobj = assemble(qc)
statevector = sv_sim.run(qobj).result().get_statevector()
plot_bloch_multivector(statevector)
```

### Comentarios:

El cubit 0 se encuentra en el estado |0> tras la aplicación del algoritmo. Esto es exactamente lo que queremos ya que al medirlo siempre colapsará a 0 probando así que el el algoritmo funciona (estábamos trabajando con una función constante).

El segundo cubit  tendrá las mismas posibilidades de colapsar tanto a 0 como a 1.

### Ejecución en backend real.

```python
# añado las mediciones para poder
qc.measure([0,1],[0,1])
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q')
```

```python
backend = least_busy(provider.backends(filters=lambda b: b.configuration().n_qubits >= 3 and
                                   not b.configuration().simulator and b.status().operational==True))
```

```python
shots = 2048
transpiled_qc = transpile(qc, backend, optimization_level=3)
qobj = assemble(transpiled_qc, shots=shots)
job = backend.run(qobj)
job_monitor(job)
```

```python
counts = job.result().get_counts()
plot_histogram(counts)
```

### Comentarios: 


Al ejecutarlo en un backend real, hay ciertas veces en las que salen resultados no esperados debido al ruido (01 y 11 no deberían aparecer teoricamente). Pese a esto, debido a ser un circuito sencillo, los resultados obtenidos son muy satisfactorios y un 98% de las veces nos devuelve los resultados esperados.
